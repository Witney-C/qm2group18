```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/Witney-C/qm2group18/main/access_colours.csv",  
            dtype={"fips":str,"colour assignment":str})

df
```

```python
df2 = pd.read_csv("/content/data/counties.csv",dtype={"fips":str})
df2
```

```python
df3 = df2.merge(df,left_on='fips',right_on='fips')
df3
```

```python
#counties with 0 clinics
zero = df3.loc[df3['colour assignment'] == '0']
zero
```

```python
#state with most counties with zero clinics
zero['state'].describe()
```

```python
#counties with lowest assignment of clinics/1000 women in demand (0-1)
low = df3.loc[df3['colour assignment'] == '0-1']
low
```

```python
#state with most counties with 0-1 clinics/1000 women
low['state'].describe()
```

```python
#counties with highest (10+) colour assignment
high = df3.loc[df3['colour assignment'] == '10+']
high
```

```python
#state with most counties with 0-1 clinics/1000 women
high['state'].describe()
```
