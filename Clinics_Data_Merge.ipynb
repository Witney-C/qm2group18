```python
data = pd.read_csv('/content/data/access_dataset.csv')
data.replace(to_replace={'u': 0},inplace=True)
data['Total publicly funded clinics, 2015 [1]'].astype(int)
data
```

```python
data2 = pd.read_csv('/content/data/county_complete.csv')
data2
```

```python
df = data[['fips', 'Total publicly funded clinics, 2015 [1]']].copy()
df
```

```python
df2 = data2[['fips', 'state','name','area_2010','unemployment_rate_2019','per_capita_income_2019']].copy()
df2
```

```python
df_joined = df2.merge(df,left_on='fips',right_on='fips')
df_joined
```

```python
df_joined.to_csv('/content/data/clinics_factors.csv')
```

```python
politics = pd.read_csv('https://raw.githubusercontent.com/tonmcg/US_County_Level_Election_Results_08-20/master/2016_US_County_Level_Presidential_Results.csv')
politics.head()
```

```python
df3 = politics[['combined_fips','per_dem','per_gop']].copy()
df3
```

```python
df_new = df_joined.merge(df3,left_on='fips',right_on='combined_fips')
df_new.drop('combined_fips', axis=1, inplace=True)
df_new
```

```python
df_new.to_csv('/content/data/clinics_factors.csv')
```
